```python
# BOOLEAN indekslash

names = np.array(['Hasan', 'Husan', 'Mirzabek', 'Elyor', 'Hasan', 'Javohir', 'Elyor'])
names
```

```python
data = np.random.randn(7, 4)
data
```

```python
names == 'Hasan'
data[names == 'Hasan'] # Hasanning barcha ma'lumotlarini ajratib oladi.
data[names != 'Hasan'] # Hasanga tegishli bo'lmagan
data[~(names == 'Hasan')] # != bilan bir xil
data[names == 'Hasan', 2:] # oxirgi 2 ta indedksni ajratadi
```

```python
mask = (names == 'Hasan') | (names == 'Husan') # "|" or (yoki)
mask
data[mask] # maskning ma'lumotlari
data[data<0] = 0 # o dan kichik bo'lsa 0 ga tenglaydi
data
```
