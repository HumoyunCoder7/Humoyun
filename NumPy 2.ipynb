```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

sns.distplot(random.normal(loc=50, scale=5, size=1000), hist=False, label='normal')
sns.distplot(random.binomial(n=100, p=0.5, size=1000), hist=False, label='binomial')

plt.show()
```

```python
# Puasson taqsimoti

from numpy import random
x = random.poisson(lam=2, size=10)
print(x)

# Vizualizatsiyasi

import matplotlib.pyplot as plt
import seaborn as sns

sns.distplot(random.poisson(lam=2, size=1000), kde=False)

plt.show()
```

```python
# Puasson va Oddiy taqsimotlar farqi

from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

sns.distplot(random.normal(loc=50, scale=7, size=1000), hist=False, label='normal')
sns.distplot(random.poisson(lam=50, size=1000), hist=False, label='poisson')

plt.show()

# Puasson va Binomial taqsimotlar farqi

sns.distplot(random.binomial(n=1000, p=0.01, size=1000), hist=False, label='binomial')
sns.distplot(random.poisson(lam=10, size=1000), hist=False, label='poisson')

plt.show()
```

```python
# Yagona taqsimot

# Har bir voqea sodir bo'lish ehtimoli teng bo'lgan ehtimollikni tavsiflash uchun ishlatiladi.
# Masalan, tasodifiy sonlarni yaratish.
# U uchta parametrga ega:
# a - pastki chegara - standart 0 .0.
# b - yuqori chegara - standart 1.0.
# size - Qaytarilgan massivning shakli.

from numpy import random
x = random.uniform(size=(2, 3))
print(x)

# Vizualizatsiya

import matplotlib.pyplot as plt
import seaborn as sns

sns.distplot(random.uniform(size=1000), hist=False)

plt.show()
```

```python
# Logistik taqsimot

# Logistik taqsimot o'sishni tavsiflash uchun ishlatiladi.
# Logistik regressiya, neyron tarmoqlar va boshqalarda mashinani o'rganishda keng qo'llaniladi.
# U uchta parametrga ega:
# loc- degani, cho'qqi qayerda. Standart 0.
# scale- standart og'ish, taqsimotning tekisligi. Standart 1.
# size- Qaytarilgan massivning shakli.

from numpy import random
x = random.logistic(loc=1, scale=2, size=(2,3))
print(x)

# Vizualizatsiaysi

import matplotlib.pyplot as plt
import seaborn as sns

sns.distplot(random.logistic(size=1000), hist=False)

plt.show()
```

```python
# Logistik va normal orasidagi farq

from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

sns.distplot(random.normal(scale=2, size=1000), hist=False, label='normal')
sns.distplot(random.logistic(size=1000), hist=False, label='logistic')

plt.show()
```

```python
# Multinomial taqsimot

# Ko'p nomli taqsimot - binomial taqsimotning umumlashtirilishi.
# U ko'p nomli stsenariylarning natijalarini tasvirlaydi, binomialdan farqli o'laroq, 
# stsenariylar ikkitadan faqat bittasi bo'lishi kerak. Masalan, populyatsiyaning qon turi, zarning o'g'irlanishi natijasi.
# U uchta parametrga ega:
# n- mumkin bo'lgan natijalar soni (masalan, zar uchun 6).
# pvals- natijalar ehtimoli ro'yxati (masalan, zar uchun [1/6, 1/6, 1/6, 1/6, 1/6, 1/6]).
# size- Qaytarilgan massivning shakli

from numpy import random
x = random.multinomial(n=6, pvals=[1/6, 1/6, 1/6, 1/6, 1/6, 1/6])
print(x)
```

```python
# Eksponensial taqsimot

# Eksponensial taqsimot keyingi hodisaga qadar vaqtni tavsiflash uchun ishlatiladi, masalan, muvaffaqiyatsizlik/muvaffaqiyat va hokazo.
# U ikkita parametrga ega:
# scale- teskari tezlik (poisson taqsimotidagi lamga qarang) sukut bo'yicha 1,0 ga teng.
# size- Qaytarilgan massivning shakli.

from numpy import random
x = random.exponential(scale=2, size=(2,3))
print(x)

# Vizualizatsiyasi

import matplotlib.pyplot as plt
import seaborn as sns

sns.distplot(random.exponential(size=7), hist=False)

plt.show()
```

```python
# Chi maydoni taqsimoti

# Chi-kvadrat taqsimoti gipotezani tasdiqlash uchun asos sifatida ishlatiladi.
# U ikkita parametrga ega:
# df- (erkinlik darajasi).
# size- Qaytarilgan massivning shakli.

from numpy import random
x = random.chisquare(df=2, size=(3,2))
print(x)

# Vizualizatsiyasi

import matplotlib.pyplot as plt
import seaborn as sns

sns.distplot(random.chisquare(df=1, size=77), hist=False)

plt.show()
```

```python
# Zipf taqsimotlari zipf qonuniga asoslangan ma'lumotlarni tanlash uchun ishlatiladi.

# Zipf qonuni: To'plamda n-chi umumiy atama eng keng tarqalgan atamaning 1/n marta.
# Masalan, ingliz tilidagi beshinchi umumiy so'z eng ko'p ishlatiladigan so'zning deyarli 1/5 qismiga to'g'ri keladi.
# U ikkita parametrga ega:
# a- tarqatish parametri.
# size- Qaytarilgan massivning shakli.

from numpy import random

x = random.zipf(a=2, size=(2, 3))
print(x)

# Vizualizatsiyasi

import matplotlib.pyplot as plt
import seaborn as sns

x = random.zipf(a=2, size=1000)
sns.distplot(x[x<10], kde=False)

plt.show()
```
